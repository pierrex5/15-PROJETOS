```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a sample DataFrame for demonstration
data = {'nota': np.random.randint(0, 11, 100),
        'turma': np.random.choice(['A', 'B', 'C', 'D'], 100),
        'idade': np.random.randint(18, 25, 100)}
df = pd.DataFrame(data)


# histograma das notas
df['nota'].plot.hist(bins=5, title='Histograma das notas')
plt.xlabel('Nota')
plt.show()

# mdia de notas por turma (barra)
media_turma = df.groupby('turma')['nota'].mean()
media_turma.plot.bar(title='Mdia de notas por turma')
plt.ylabel('Mdia da nota')
plt.show()

# scatter nota x idade
df.plot.scatter(x='idade', y='nota', title='Nota vs Idade')
plt.show()
```
