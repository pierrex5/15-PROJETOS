Para demonstrar o `pd.read_csv()`, vamos primeiro criar um arquivo CSV simples e depois l√™-lo.

```python
import pandas as pd
import os

# Conte√∫do de exemplo para o arquivo CSV
data = {
    'Produto': ['Ma√ß√£', 'Banana', 'Laranja', 'Uva'],
    'Preco': [2.50, 3.00, 4.20, 5.80],
    'Quantidade': [100, 150, 75, 200]
}
df_sample = pd.DataFrame(data)

# Caminho para o arquivo CSV tempor√°rio
csv_path = '/tmp/frutas.csv'

# Salva o DataFrame como um arquivo CSV
df_sample.to_csv(csv_path, index=False)

print(f"Arquivo '{csv_path}' criado com sucesso!")
```

```python
import pandas as pd

# L√™ o arquivo CSV em um DataFrame
df_lido = pd.read_csv('/tmp/frutas.csv')

# Exibe o DataFrame lido
print("Conte√∫do do DataFrame lido:")
display(df_lido)
```

### üü¶ 7. Sele√ß√£o de dados
- `df['coluna']` ‚Üí seleciona uma coluna.
- `df[['col1','col2']]` ‚Üí seleciona v√°rias colunas.
- `df.loc[linha]` ‚Üí busca por √≠ndice.
- `df.iloc[linha]` ‚Üí busca por posi√ß√£o.

```python
print("Selecionando a coluna 'Produto' de df_lido:")
display(df_lido['Produto'])

print("\nSelecionando as colunas 'Produto' e 'Preco' de df_lido:")
display(df_lido[['Produto', 'Preco']])

print("\nSelecionando a linha com √≠ndice 0 (df.loc[0]) de df_lido:")
display(df_lido.loc[0])

print("\nSelecionando a linha na posi√ß√£o 1 (df.iloc[1]) de df_lido:")
display(df_lido.iloc[1])
```

### üü¶ 8. Filtros
- `df[df['coluna'] > valor]`

```python
print("Filtrando df_lido para produtos com 'Preco' maior que 3.50:")
display(df_lido[df_lido['Preco'] > 3.50])
```

### üü¶ 9. Ordena√ß√£o
- `df.sort_values('coluna')`

```python
print("Ordenando df_lido por 'Preco' em ordem crescente:")
display(df_lido.sort_values('Preco'))

print("\nOrdenando df_lido por 'Quantidade' em ordem decrescente:")
display(df_lido.sort_values('Quantidade', ascending=False))
```
